```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the datasets
```

```python
orders = pd.read_csv(r"C:\Users\USER\Downloads/orders.csv")

order_details = pd.read_csv(r"C:\Users\USER\Downloads/order_details.csv")

pizzas = pd.read_csv(r"C:\Users\USER\Downloads/pizzas.csv")

pizza_types = pd.read_csv(r"C:\Users\USER\Downloads/pizza_types.csv", encoding='Unicode_escape')
```

```python
#Merge into one dataframe
df = pd.merge(orders, order_details, on='order_id')

df = pd.merge(df, pizzas, on='pizza_id')

df = pd.merge(df, pizza_types, on='pizza_type_id')
```

```python
df.head()
```

```python
df.info()
```

```python
#Ensure date and time columns are in datetime format

df['date'] = pd.to_datetime(df['date'])

df['time'] = pd.to_datetime(df['time'])
```

```python
df.info()
```

```python
#To check for total revenue or sales

df["revenue"] = df["price"] * df["quantity"]
```

```python
#Q1: Total Revenue/Sales

total_revenue = df["revenue"].sum()
print(f"1. Total Revenue: ${total_revenue: ,.2f}")
```

```python
#Q2: Find the total quantity sold

total_quantity = df["quantity"].sum()
print(f"2 Total Quantity Sold: {total_quantity:,}")
```

The total revenue for the year was $ 817,860.05, driven by 49,574 units sold. This provides a baseline for evaluating the store's annual performance and scaling potential.

```python
#Q3 Find the total orders

total_orders = df["order_id"].nunique()
print(f"3 Total Orders: {total_orders:,}")
```

```python
#Q4 How many pizza types do they sell

pizza_type = df["name"].nunique()
print(f"4 Total Pizza Types: {pizza_type}")
```

```python
#Q5 Find the average price of the pizzas

avg_pizza_price = df["price"].mean()
print(f"5 Average Price of Pizzas: {avg_pizza_price: .2f}")
```

With 49,574 unique orders, we can calculate that the average customer spend is approximately 16.50. The average menu price of $16.49 suggests a mid-range pricing strategy


```python
#Q6 What are the peak hours of sales

df['hour'] = df['time'].dt.hour

peak_hours = df.groupby('hour')['order_id'].nunique().sort_values(ascending=False)

print("\n 6. The Peak Hour of Sales (Top 3): ")
print(peak_hours.head(3))
```

```python
#Q7 Total sales made on each day of the week

df['day_of_week'] = df['date'].dt.day_name()

sales_by_day = df.groupby('day_of_week')['revenue'].sum().sort_values(ascending=False)

print("7. Sales by Day of the Week:")


print(sales_by_day)
```

```python
#Q8 Top 5 best selling pizzas (by quantity) 

top_5_pizzas = df.groupby('name')['quantity'].sum().sort_values(ascending=False)

print("5. The top 5 best selling pizzas are:")

print(top_5_pizzas.head())
```

```python
#Q9 Sales made in each month

df['month']= df['date'].dt.month_name()

sales_in_each_month = df.groupby('month')['revenue'].sum().sort_values(ascending=False)

sales_in_each_month
```

### Insight: Monthly Sales Trends
The analysis reveals that the pizza place maintains a steady stream of income throughout the year, with a narrow variance between the highest and lowest-earning months.

1. **Top Performance:** July is the most profitable month, followed closely by May. This may correlate with social gatherings or events during the summer season.

2. **Lowest Performance:** Sales dip slightly in September and October.

3. **Observation:** The differences between the highest month (July) and the lowest month (October) is only about $8,500, showing that the business is not heavily seasonal and enjoys stable demand year-round.

```python
#Q10 Pizzas not doing well (bottom 5 by quantity)

bottom_5_pizzas = df.groupby('name')['quantity'].sum().sort_values(ascending=True).head(5)

print("\nBottom 5 Pizzas (Potential Menu Removal):")
print(bottom_5_pizzas)
```
